```python
#11.Создайте еще одни столбец в датафреме с именем «good_temp» 
#и запишите в него результат сравнения текущей температуры со средним значением (True or False).
data['good_temp'] = data['temp'].map(lambda f: f > data['temp'].mean())
```

```python
#15 Выведите датафрейм, отсортированный по полю count, в обратном порядке.
data.sort_values('count', ascending=False)
```

```python
#16. Создайте еще одни датафрем на основе описания поля 
#season (1 = spring, 2 = summer, 3 = fall, 4 = winter).
season = pd.DataFrame(
    [
        [1, 'spring'],
        [2, 'summer'],
        [3, 'fall'],
        [4, 'winter']
    ],
    columns = ['id', 'lable']
)
season
```

```python
#17. Соедините исходный датафрем с даатфремом, 
#созданным на предыдущем шаге по указанному выше правилу. Выведите полученный результат.
data = pd.merge(season, data, left_on = 'id', right_on = 'season')
data.head()
```

```python
#18. Сделайте группировку по временам года и количеству (count).
data[['lable', 'count']].groupby('lable').count()
```

```python
#19. Создайте pivot_table по любым столбцам и опишите полученный результат.
data.pivot_table('count',['lable','temp'])
```

Вывели количество арендуемых помещений (count) в зависимости от ыремени года и температуры

```python
#20. Получите процентное соотношение count в зависимости от времен года.
d = data.pivot_table('count','lable')
d.div(d.sum(0), axis='columns')
```
