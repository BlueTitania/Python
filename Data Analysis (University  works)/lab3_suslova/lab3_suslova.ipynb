Чтение даных 
--  
  Для чтения используем конструкцию **open**  
  В ней указываем имя файла (user_ids.txt), параметр чтения ('r') и обозначаем объект файла как f.  
    'w' - очистить файл для записи в него новых данных.
    'a' - открыть файл для добавления новых записей в конец.   
       
  Воспользуемся методом strip(). Он убирает с начала и конца строки переносы строк.

```python
with open('user_ids.txt', 'r') as f:
    for line in f:
        print(line.strip())
```

```python
#Запишем очередной ID в лист 
user_ids = []
with open( 'user_ids.txt', 'r' ) as f:
    for line in f:
        user_id = line.strip()        
        user_ids.append( user_id )
```

```python
#Выведите на экран первые 5 элементов листа user_ids.
user_ids[:5]
```

```python
#. Какой ID будет в этом наборе последним?
print('Last elem = ', user_ids[5])
```

```python
#Сколько всего элементов в этом листе? 
len(user_ids)
```

```python
#Подсчет уникальных ID
unique_ids = set( user_ids )
print( 'Уникальных ID в файле {}'.format( len( unique_ids ) ) )
```

```python
#Какое количество уникальных ID в файле data.txt?
idhs = []
with open('data.txt', 'r') as f:
    for line in f:
        idh = line.strip()
        idhs.append(idh)
print( 'Уникальных ID в файле {}'.format(len(set(idhs))))
```

Чтение файлов с заголовками
--

```python
#Какая строчка появится первой при чтении файла user_ids_headers.txt описанным выше способом?
with open( 'user_ids_headers.txt', 'r' ) as f:
    for line in f:
        print( line.strip())
        break
```

```python
Started = True
with open( 'user_ids_headers.txt', 'r' ) as f:
    for line in f:
        # запись "if Started" эквивалентна "if Started == True"
        if Started:
            # эта строчка выполняется один раз только на первом шаге цикла
            Started = False
        else:
            # эту строчку выполняем со второго и во всех следующих шагах цикла
            print( line.strip() )
```

```python
#альтернатива
with open( 'user_ids_headers.txt', 'r' ) as f:
    for line in f.readlines()[1:]:
        print(line.strip())
```

Для разделения строки line на столбцы используется метод split. В качестве аргумента в нее подставляется разделитель столбцов. 

```python
with open( 'data_3_columns.txt', 'r' ) as f:
    for line in f:
        line = line.strip().split('\t')
        print( line )
```

Приведение к числу

```python
with open( 'data_3_columns.txt', 'r' ) as f:
    for line in f:
        line = line.strip().split('\t')
        medium = line[0]
        source = line[1]
        amount_paid = float(line[2].replace(',', '.'))
        print( source, medium, amount_paid )
```

Подсчет суммы в файлах
--

```python
total_sum = 0
with open( 'data_3_columns.txt', 'r' ) as f:
    for line in f:
        line = line.strip().split('\t')
        amount_paid = float( line[2].replace(',', '.') )
        total_sum += amount_paid
        print( 'Текущая сумма расходов: {:.2f}'.format( total_sum ) )
```

```python
#Какое значение суммы получаем в последнем примере?
total_sum
```

```python
#Сумма для источников google и yandex(seo)
sum_google = sum_yandex = 0
with open( 'data_3_columns.txt', 'r' ) as f:
    for line in f:
        line = line.strip().split('\t')
        line[2] = float(line[2].replace(',','.'))
        if line[1] == 'google' :
            sum_google += line[2]
        elif line[1] == 'yandex' and line[0] == 'seo':
            sum_yandex += line[2]
print( 'Текущая сумма расходов google: {:.2f}'.format( sum_google ) )
print( 'Текущая сумма расходов yandex: {:.2f}'.format( sum_yandex ) )
```

Библиотека openpyxl

```python
import openpyxl

workbook = 'data.xlsx'
sheetName = 'data'
sheet = openpyxl.load_workbook(workbook).get_sheet_by_name(sheetName)
for line in sheet['A1:I1']:
    print(line)
    print(line[0].value)
    #Какое значение у ячейки E1?
    print(sheet["E1"].value)
```

Чтение содержимого Excel-файла в цикле

```python
for line in sheet[ 'E2:I5' ]:
    medium = line[0].value
    source = line[1].value
    amount_paid = line[4].value
    print( source, medium, amount_paid )
type(amount_paid)
```
