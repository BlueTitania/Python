```python
with open( '/home/lorer/Загрузки/data_3_columns.txt', 'r' ) as f:
    summg = summy = 0
    for line in f:
        line = line.strip().split('\t')
        line[2] = float(line[2].replace(',','.'))
        if line[1] == 'google':
            summg += line[2]
        elif line[1] == 'yandex' and line[0] == 'seo':
            summy += line[2]
    print("total summ for google: " + str(summg))
    print("total summ for yandex seo: " + str(summy))
```
