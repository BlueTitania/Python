```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('churn.csv', delimiter  = ',')
df.head(5)
```

```python
drop_col = []
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    if pct_missing == 1:
        drop_col.append(col)
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.drop(columns=drop_col)
```

```python
for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        print(df[col].describe())
```

```python
for col in df.columns:
    # if df[col].dtype == 'int64' or df[col].dtype == 'float64':
    #if col in ['accountLevel']:#, 'hardBalance', 'purchaseCount', 'realWorldCurrencySpent'] :
        #МКР это разница между 75% и 25% квартилем. 
        #Значения 25% перцентили минус 1.5 х МКР
        #Значения 75% перцентили плюс 1.5 х МКР
        #too_big = np.mean(df[col] > df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25)))
        #too_small = np.mean(df[col] < df[col].quantile(0.25) - 1.5* (df[col].quantile(0.75) - df[col].quantile(0.25)))
        #print('Колонка {} - имеет {}% выбросов'.format(col, (too_big+too_small)*100))
    if col in ['Preset3heroLevel']:
        too_big = np.mean(df[col] > 12)
        print('Колонка {} - имеет {}% выбросов'.format(col, (too_big)*100))
    if col in ['appActiveTime']:
        too_small = np.mean(df[col] < 0)
        print('Колонка {} - имеет {}% выбросов'.format(col, (too_small)*100))
    if col in ['mmr']:
        too_small = np.mean(df[col] > 9999)
        print('Колонка {} - имеет {}% выбросов'.format(col, (too_small)*100))
```

```python
for col in df.columns:
#     if col in ['softBalance', 'hardBalance', 'purchaseCount', 'realWorldCurrencySpent'] :
#         df = df[df[col] > df[col].quantile(0.25) - 1.5* (df[col].quantile(0.75) - df[col].quantile(0.25))] 
#         df = df[df[col] < df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25))]
    if col in ['Preset3heroLevel']:
        df = df[df[col] < 12]
    if col in ['appActiveTime']:
        df = df[df[col] > 0]
    if col in ['mmr']:
        df = df[df[col] < 9999]
```

```python
df
```

```python
# Список признаков с высоким процентом одинаковых значений
num_rows = len(df.index)
low_information_cols = [] #

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.95:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```

```python
# уберем не полезные данные
df.drop(['battlepassElite', '_col21', 'clanpassElite', 'rn'], axis='columns', inplace=True)
```

```python
for index, row in df.iterrows():
    if (row['Preset1hero'] != 'ROOKIE' 
        and (
            row['Preset1hero'] == row['Preset2hero'] 
            and row['Preset1hero'] == row['Preset3hero']
            and row['Preset2hero'] == row['Preset3hero']
        )
    ):
        df = df.drop(index)
```
