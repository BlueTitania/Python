```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('82820ca0-c1bf-4e63-9895-8991b1b27554.csv', delimiter  = ',')
df.head(5)
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        print(df[col].describe())
```

```python
for col in df.columns:
    # if df[col].dtype == 'int64' or df[col].dtype == 'float64':
    if col in ['softBalance', 'hardBalance', 'purchaseCount', 'realWorldCurrencySpent'] :
        # Выбросы считаем все что нахоится за границами 3х сигма
        too_big = np.mean(df[col] >  df[col].mean() + 3*df[col].std())
        too_small = np.mean(df[col] < df[col].mean() - 3*df[col].std())
        print('Колонка {} - имеет {}% выбросов'.format(col, round((too_big+too_small)*100)))
```

```python
for col in df.columns:
    if col in ['softBalance', 'hardBalance', 'purchaseCount', 'realWorldCurrencySpent'] :
        
        too_big = np.mean(df[col] >  df[col].mean() + 3*df[col].std())
        too_small = np.mean(df[col] < df[col].mean() - 3*df[col].std())
        print(col, df[col].mean(),  3*df[col].std(),  3*df[col].std(), df[col].mean() - 3*df[col].std(), df[col].mean() + 3*df[col].std())
        if round((too_big+too_small)*100) > 0:
            df = df[df[col] > df[col].mean() - 3*df[col].std()] 
            df = df[df[col] < df[col].mean() + 3*df[col].std()]
```

```python
df
```

```python
# Список признаков с высоким процентом одинаковых значений
num_rows = len(df.index)
low_information_cols = [] #

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.95:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```

```python
# уберем не полезные данные
df.drop(['battlepassElite', '_col21', 'clanpassElite', 'rn'], axis='columns', inplace=True)
```
